# Label Encoding Thorugh Pandas

```python
# import required libraries
import pandas as pd
import numpy as np
```

```python
# creating initial dataframe
bridge_types = ('Arch','Beam','Truss','Cantilever','Tied Arch','Suspension','Cable')
df = pd.DataFrame(bridge_types, columns=['Bridge_Types']) # converting type of columns to 'category'
df
```

```python
df['Bridge_Types'] = df['Bridge_Types'].astype('category') # Assigning numerical values and storing in another column
df['Bridge_Types_Cat'] = df['Bridge_Types'].cat.codes
df
```

# Label Encoding Thorugh Scikit-Learn

```python
from sklearn.preprocessing import LabelEncoder 
# creating initial dataframe
df = pd.DataFrame(bridge_types, columns=['Bridge_Types']) # creating instance of labelencoder
df
```

```python
labelencoder = LabelEncoder() # Assigning numerical values and storing in another column
df['Bridge_Types_Cat'] = labelencoder.fit_transform(df['Bridge_Types'])
df
```

# One Hot Encoding Thorugh Scikit-Learn

```python
from sklearn.preprocessing import OneHotEncoder # creating instance of one-hot-encoder
enc = OneHotEncoder(handle_unknown='ignore') # passing bridge-types-cat column (label encoded values of bridge_types)
enc_df = pd.DataFrame(enc.fit_transform(df[['Bridge_Types_Cat']]).toarray()) # merge with main df bridge_df on key values
df = df.join(enc_df)
df
```

# One Hot Encoding Thorugh Pandas

```python
df = pd.DataFrame(bridge_types, columns=['Bridge_Types'])# generate binary values using get_dummies
df
```

```python
dum_df = pd.get_dummies(df, columns=["Bridge_Types"], prefix=["Type_is"] )# merge with main df bridge_df on key values
df = df.join(dum_df)
df
```
